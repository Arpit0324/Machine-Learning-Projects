# Outlier Handling

```python
plt.figure(figsize = (15,10))

ax = plt.subplot(221)
plt.boxplot(placement_prediction['ssc_p'])
ax.set_title('Secondary School Percentage')

ax = plt.subplot(222)
plt.boxplot(placement_prediction['hsc_p'])
ax.set_title('Higher secondary Percentage')

ax = plt.subplot(223)
plt.boxplot(placement_prediction['degree_p'])
ax.set_title('UG Percentage')

ax = plt.subplot(224)
plt.boxplot(placement_prediction['etest_p'])
ax.set_title('Employability Percentage')
```

```python
Q1 = placement_prediction['hsc_p'].quantile(0.25)
Q3 = placement_prediction['hsc_p'].quantile(0.75)
IQR = Q3 - Q1

filter = (placement_prediction['hsc_p'] >= Q1 - 1.5 * IQR) & (placement_prediction['hsc_p']<= Q3+ 1.5*IQR)
placement_filtered= placement_prediction.loc[filter]
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder

object_cols= ['gender','workex','specialisation','status']

label_encoder = LabelEncoder()

for col in object_cols:
    placement_filtered[col]= label_encoder.fit_transform(placement_filtered[col])
    
placement_filtered.head(10)
```
